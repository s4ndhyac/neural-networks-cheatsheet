### 1: Python & Data Exploration

We load the .txt file provided. And set the variable Y as the target value/label as the last column. We set all other columns except the last column as feature columns in variable X.

```python
#Loading the data
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(0)
iris = np.genfromtxt("data/iris.txt",delimiter=None)
Y = iris[:,-1] 
X = iris[:,0:-1]
```

1. 1. Use X.shape to get the number of features and the data points. Report both numbers, mentioning which
number is which.

```python
#1. Get no. of features and data points
print('The no. of features is: {0}'.format(X.shape[1]))
print('The no. of data points are: {0}'.format(X.shape[0]))
```

2. For each feature, plot a histogram ( plt.hist ) of the data values 

We use the matplotlib library to generate plots inline. Since the no. of features are 4 as seen above, for each feature i from 0 to 3 both inclusive, we plot a histogram using plt.hist and label it with the feature.

```python
# 2. Plot a histogram of values for each feature
for i in range(0, 4):
    plt.hist(X[:, i], label='feature_{}'.format(i+1))
    plt.legend()
    plt.show()
    plt.close()
```

3. Compute the mean & standard deviation of the data points for each feature ( np.mean , np.std )

For each feature from 0 to 3 both inclusive, we use the numpy libraries, np.mean and np.std functions to calcualte the mean and standard deviation.

```python
# 3. Plot mean and standard deviation of each feature
for i in range(0, 4):
    print('The Mean of feature {} is {}'.format(i, np.mean(X[:, i])))
    print('Standard deviation of feature {} is {}'.format(i, np.std(X[:, i])))
```

4. For each pair of features (1,2), (1,3), and (1,4), plot a scatterplot (see plt.plot or plt.scatter ) of the
feature values, colored according to their target value (class). (For example, plot all data points with y = 0
as blue, y = 1 as green, etc.)

We store the no. of data points in the dataPoints variable. We need to generate plots for features 2,3,4 against 1, so we store the indexes of these features 1,2,3 in an array. Then for each feature, for each datapoint, we assign a color (we know there are 3 classes from the question description) - blue for class y=0, green for class y=1 and red for class y=2 and plot the point on the graph, with feature_1 always on the x-axis and feature_2, feature_3, feature_4 on the y-axis in turns.

```python
# 4. Plot a scatter plot for each pair of features colored according to their target value
dataPoints = X.shape[0]
features = [1, 2, 3]
for feature in features:
    for i in range(0, dataPoints):
        if Y[i] == 0:
            plt.scatter(X[i, 0], X[i, feature], c='b')
        if Y[i] == 1:
            plt.scatter(X[i, 0], X[i, feature], c='g')
        if Y[i] == 2:
            plt.scatter(X[i, 0], X[i, feature], c='r')
    plt.xlabel('feature_1')
    plt.ylabel('feature_{}'.format(feature+1))
    plt.legend()
    plt.show()
    plt.close()
```
