### 1: Linear Regression
1. We load the data from the curve80.txt file. The code expects data in the (M, N) shape and so we make sure it's 2-dimensional by using np.atleast_2d(). The first column is the scalar features value and the second column is the target value y for each value. We then split it into the train and test datasets in the ratio 75:25. The Y test and Y train datasets are also made into a 2-d matrix using np.atleast_2d()

```python
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#load data
data = np.genfromtxt("data/curve80.txt",delimiter=None)
#features column
X = data[:, 0] 
# making X 2-dimensional
X =  np.atleast_2d(X).T
#target values
Y = data[:, 1]
#splitting X and Y in the ratio 75:25 between train and test datasets
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y,0.75)
# Making Ytr and Yte 2-dimensional
Ytr = np.atleast_2d(Ytr).T
Yte = np.atleast_2d(Yte).T
# Printing shapes of Xtr, Xte, Ytr and Yte
print(Xtr.shape, Xte.shape, Ytr.shape, Yte.shape)
```

2.a lr is a linear regression predictor of y given x. xs denotes the large no. of sample data points. ys is the prediction made by the linear regression predictor lr at xs. The training data points are plotted as a scatter plot in blue. The predictor line is plotted in black.

```python
# lr - the linear regression predictor of y given x
lr = ml.linear.linearRegress(Xtr, Ytr)
# evenly spaced data points between 0 and 10, sampled 200 times
xs = np.linspace(0, 10, 200)
# Making xs a Mx1 matrix
xs = xs[:, np.newaxis]
# make predictions at xs
ys = lr.predict(xs)

# Make a scatter plot of X and Y train data points in blue
plt.scatter(Xtr, Ytr, c = 'blue')
# Plot the linear regression predictor line in black
plt.plot(xs, ys, c= 'black')
plt.title('Regression Function')
plt.show()
```

b. We use `lr.theta` to get the regression coefficients. We note that the regression coefficients closely match the plots.

```python
# Printing the linear regression coefficient
print ('The Regression Coefficients are = ', lr.theta)
```

c. We get the predictions made on X train and X test. And calculate the MSE on the training and test data using the formula -> np.mean((Y train predicted - Y train)**2)

```python
YtrPred = lr.predict(Xtr)
YtePred = lr.predict(Xte)
#calculating and printing the mean squared error on the Training data
mseTrain = np.mean((YtrPred - Ytr) ** 2)
print ("The Mean Square Error on Training Data is = ", mseTrain)
#calculating and printing the mean squared error on the Test data
mseTest = np.mean((YtePred - Yte) ** 2)
print ("The Mean Square Error on Test Data is = ", mseTest)
```

3.a. For each degreee in the degrees array, we call the `ml.transforms.fpoly` function to add features to the polynomial of given degree. We then use the rescale function to rescale the features again since it is possible that the features may include very large values too. We then train the model on the recaled feature matrix. Similarly, we do polynomial expansion and feature rescaling on the test and sample datasets too before, predicting via the lr predictor and plotting the functions. We also store the MSE errors on train and test data for each degree polynomial function to be used later.

```python
# Array to store the degrees of polynomials, trainingErrors and testErrors for plotting
degrs = [1, 3, 5, 7, 10, 18]
trError = []
teError = []
for deg in degrs:
    
    # Adding polynomial features and rescaling train, test and sample data points
    XtrP = ml.transforms.fpoly(Xtr, deg, bias=False)
    XtrP, params = ml.transforms.rescale( XtrP )
    lr = ml.linear.linearRegress( XtrP, Ytr )
    
    XteP,_ = ml.transforms.rescale( ml.transforms.fpoly(Xte, deg, False ), params )
    xsP,_ = ml.transforms.rescale( ml.transforms.fpoly(xs, deg, False ), params )

    # Predict on the sample data points after polynomial expansion and rescaling using the linear regression predictor
    ys = lr.predict(xsP) # make predictions at xs
    plt.scatter(Xtr, Ytr, c = 'blue')
    axs = plt.axis()
    plt.plot(xs, ys, c = 'black')
    plt.axis(axs)
    plt.title(str(deg) + " degree polynomial function")
    plt.show()
    
    # Calculate the MSE in test and train datasets and append to the trError and teError arrays to be used for plots later
    trPredMSError = lr.mse(XtrP, Ytr)
    tePredMSError = lr.mse(XteP, Yte)
    trError.append(trPredMSError)
    teError.append(tePredMSError)
```

b. We use the train and test MSE errors stored in arrays in the previous step and plot semilogy plots with the train/test error on the y-axis and the polynomial degree on the x-axis. The Training error is in blue and the Test error is in red.

```python
plt.semilogy(degrs, trError, c = 'red')
plt.semilogy(degrs, teError, c = 'blue')
plt.xticks(degrs, degrs)
plt.title("Training and Test Error vs Degree of Polynomial")
plt.xlabel('Degree of polynomial')
plt.show()
```

c. From the plot in b. above we can see that when the degree of the polynomial is 10, the MSE is least for that value. Therefore, I would recommend the polynomial of degree 10.

### Problem 2: Cross-Validation

1. For each degree in the degrs array, we calculate the test MSE as done previously in problem 1. We calculate n-Fold cross-validation MSE by calculating the MSE for each value in the range of the nFold 5 in this case and then taking a mean/average of the MSEs. We add the MSE thus calculated to the n-Fold cross-validation MSE array. We plot the n-Fold cross-validation MSEs in Red and the Test MSEs in blue against the degree of the polynomial in the x-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
import mltools as ml
%matplotlib inline

data = np.genfromtxt("data/curve80.txt",delimiter=None)
X = data[:, 0]
X = X[:,np.newaxis]
Y = data[:, 1]
Xtr,Xte,Ytr,Yte = ml.splitData(X,Y,0.75)

degrs = [1, 3, 5, 7, 10, 18]

testMSEs = []
for deg in degrs: 
    # Adding polynomial features and rescaling train, test and sample data points
    XtrP = ml.transforms.fpoly(Xtr, deg, bias=False)
    XtrP, params = ml.transforms.rescale( XtrP )
    lr = ml.linear.linearRegress( XtrP, Ytr )    
    XteP,_ = ml.transforms.rescale( ml.transforms.fpoly(Xte, deg, False ), params )
    # Calculate the MSE in test dataset and append to the teError array to be used for a plot later
    predMSError = lr.mse(XteP, Yte)
    testMSEs.append(predMSError)

nFolds = 5;
validMSEs = []
for deg in degrs:
    J = []
    for fold in range(nFolds):
        # using the ith fold in the range as validation
        Xti,Xvi,Yti,Yvi = ml.crossValidate(Xtr,Ytr,nFolds,fold)
        Yvi = Yvi[:, np.newaxis]
        XtiP = ml.transforms.fpoly(Xti, deg, bias=False)
        XtiP, params = ml.transforms.rescale( XtiP )
        lr = ml.linear.linearRegress( XtiP, Yti )
        XviP,_ = ml.transforms.rescale( ml.transforms.fpoly(Xvi, deg, False ), params )
        predMSError = lr.mse(XviP, Yvi)
        J.append(predMSError)
    validMSEs.append(np.mean(J))

plt.semilogy(degrs, validMSEs, c = 'red')
plt.semilogy(degrs, testMSEs, c = 'blue')
plt.xticks(degrs, degrs)
plt.xlabel('Degree of Polynomial')
plt.title('Average Cross Validation Error, Test Error vs Degree of Polynomial')
plt.show()
```

2. From the plot in 1. we can observe that the five-Fold cross-validation MSE is less than the test MSE at all points upto a polynomial of degree about 7, at which point the plots intersect and then five-fold cross-validation MSE is higher than the test MSE. The 5-fold cross-validation MSE decreases slightly till a polynomial of degree 5 and increases slightly from 5 till degree 10, post which it increases steeply. Whereas the test MSE remains, almost constant upto a polynomial of degree 10 post which it increases steadily. 

3. From the plot in 1. we can observe that the five-Fold cross-validation MSE decreases for a polynomial of degree 1 till 5. It has it's least value for a polynomial of degree 5 after which it increases slightly till degree 10 and then steeply increases for a polynomial of degree 18. Therefore, based on the 5-fold cross-validation MSE, I would recommend a polynomial degree of value 5.

4. For each value of the fold in the nFolds array, for the chosen value of degree of polynomial 5 in the problem 3. we calculate the n-fold cross-validation MSE as we did in problem 1. and store it in array validMSEs and plot it in red later. We observe that, for a polynomial of degree 5, the n-fold cross-validation MSE initially decreases from n=2 to 3, falses below 1, it then increases to just above 1 at n=4 and then it decreases to just below 1 at n=5 and steadily decreases for n=6, 10, 12, 15 to reach a value of ~0.8 at n=15. 
This pattern implies that the cross-validation error initially increases and then steadily decreases with the increase in the no. of folds. The cross-validation error is inversely proportional to the increase in the no. of folds.


```python
deg = 5
nFolds = [2,3,4,5,6,10,12,15];
validMSEs = []
J = []
for nFold in nFolds:
    # for each value of nFold in the nFolds array
    for fold in range(nFold):
        # using the ith fold in the range as validation
        Xti,Xvi,Yti,Yvi = ml.crossValidate(Xtr,Ytr,nFold,fold)
        Yvi = Yvi[:, np.newaxis]
        XtiP = ml.transforms.fpoly(Xti, deg, bias=False)
        XtiP, params = ml.transforms.rescale( XtiP )
        lr = ml.linear.linearRegress( XtiP, Yti )
        XviP,_ = ml.transforms.rescale( ml.transforms.fpoly(Xvi, deg, False ), params )
        predMSError = lr.mse(XviP, Yvi)
        J.append(predMSError)
    validMSEs.append(np.mean(J))

plt.semilogy(nFolds, validMSEs, c = 'red')
plt.xticks(nFolds, nFolds)
plt.xlabel('No. of Folds')
plt.title('Average Cross Validation Error vs No. of Folds')
plt.show()
```
